```python
runtime_medians = movies_revs[movies_revs['runtime'] != 0]['runtime'].median()
movies_revs['runtime_for_plot'] = movies_revs['runtime'].replace(0, runtime_medians)
```

```python
(movies_revs['runtime_for_plot'] == 0).sum()
```

```python
#Since there are movies with no voting averages, instead of filling them with median which could be misleading for our analysis, we filter the rows.
(movies_revs['vote_average'] == 0.000).sum()
```

```python
movies_cleaned = movies_revs[movies_revs['vote_average'] > 0 ]
(movies_cleaned['vote_average'] == 0.000).sum()
```

```python
#Now since we replaced the '0's with median values for runtime and filtered vote_averages, lets plot a scatterplot to see if 'runtime'has any effect on 'voting average'.

plt.figure(figsize = (10,6))
sns.scatterplot(data=movies_cleaned, x ='runtime_for_plot', y='vote_average', alpha =0.5)
plt.title('Does runtime affect ratings?')
plt.xlabel('runtime')
plt.ylabel('vote_average')
plt.grid(True)
plt.show()
```

```python
#Creating a top movies data frame which we might use later.
top_rating = movies_cleaned['vote_average'].max()
top_movies = movies_cleaned[movies_cleaned['vote_average'] == top_rating]
top_movies[['title', 'vote_average', 'popularity', 'runtime']].head()
```

```python
#Grouping runtime to vote_average, whilst taking the average of the runtime column to have a clean plot

#Since the graph is still cluttered, I am rounding the voting average
movies_cleaned.loc[:,'rounded_averages'] = movies_cleaned['vote_average'].round(1)

movies_cleaned.groupby('rounded_averages')['runtime'].mean().sort_index().plot(kind='line', marker= 'o', figsize = (10,5))
plt.title('Average runtime per Vote average')
plt.xlabel('Vote Average')
plt.ylabel('Runtime average (mins)')
plt.grid(True)
plt.show()
```

```python
# Temporary fix for plotting
# Create a temporary copy for EDA
tmdb_temp = movies_cleaned.copy()
```

```python
# Safely parse stringified lists into actual Python lists
tmdb_temp['genres'] = tmdb_temp['genres'].apply(
    lambda x: ast.literal_eval(x) if isinstance(x, str) else (x if isinstance(x, list) else [])
)

tmdb_temp['actors'] = tmdb_temp['actors'].apply(
    lambda x: ast.literal_eval(x) if isinstance(x, str) else (x if isinstance(x, list) else [])
)

# Extract main_genre (first genre in the list), fill NaNs with placeholder
tmdb_temp['main_genre'] = tmdb_temp['genres'].apply(
    lambda x: x[0] if isinstance(x, list) and len(x) > 0 else np.nan
)
if tmdb_temp['main_genre'].mode().empty:
    tmdb_temp['main_genre'].fillna("Unknown", inplace=True)
else:
    tmdb_temp['main_genre'].fillna(tmdb_temp['main_genre'].mode()[0], inplace=True)

# Extract main_actor (first actor in the list), fill NaNs with placeholder
tmdb_temp['main_actor'] = tmdb_temp['actors'].apply(
    lambda x: x[0] if isinstance(x, list) and len(x) > 0 else np.nan
)
if tmdb_temp['main_actor'].mode().empty:
    tmdb_temp['main_actor'].fillna("Unknown", inplace=True)
else:
    tmdb_temp['main_actor'].fillna(tmdb_temp['main_actor'].mode()[0], inplace=True)

# Count number of genres per movie
tmdb_temp['num_genres'] = tmdb_temp['genres'].apply(
    lambda x: len(x) if isinstance(x, list) else 0
)

# Count number of actors per movie
tmdb_temp['num_actors'] = tmdb_temp['actors'].apply(
    lambda x: len(x) if isinstance(x, list) else 0
)

# Quick sanity check
print(tmdb_temp['main_genre'].value_counts(dropna=False).head())
print(tmdb_temp['main_actor'].value_counts(dropna=False).head())
```

```python
#Creating a correlation matrix 
corr_matrix = tmdb_temp.drop(columns=['adult']).corr(numeric_only=True)
plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix,annot=True, fmt ='.2f',cmap='coolwarm',linewidth=0.5)

plt.title('Correlation Matrix for numeric variables')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```

```python
#Checking Distributions for key features like, 'vote_average','runtime','popularity','vote_count'

#Creating 2x2 subplots
fig, axs= plt.subplots(2,2,figsize=(12,8))
#Distribution of vote_average
sns.histplot(tmdb_temp['vote_average'],bins=20, kde= True, ax=axs[0,0])
axs[0,0].set_title('Distribution of Vote Average')
#Distribution of runtime
sns.histplot(tmdb_temp['runtime'], bins=20, kde= True, ax= axs[0,1])
axs[0,1].set_title('Distribution of runtime')
#Distribution of popularity
sns.histplot(tmdb_temp['popularity'], bins=20, kde= True, ax= axs[1,0])
axs[1,0].set_title('Distribution of Popularity')
#Distribution of vote_count
sns.histplot(tmdb_temp['vote_count'], bins=20, kde= True, ax= axs[1,1])
axs[1,1].set_title('Distribution of Vote Count')
plt.tight_layout()
plt.show()
```

#EDA Summary
In this notebook, I performed exploratory data analysis on the TMDB movie dataset to understand feature distributions and relationships with the target variable vote_average. I created temporary columns such as main_genre, main_actor, num_genres, and num_actors for visualization purposes and handled missing values with mode fills only for plotting. I also filtered out movies with vote_average == 0 to avoid skewed visualizations. For scatterplots, I temporarily replaced anomalous runtime values with the median to inspect patterns. All transformations were confined to a temporary dataframe (tmdb_temp) to preserve the integrity of the original data. No changes were saved to movies_cleaned.csv; all permanent feature engineering and imputation steps are deferred to the Preprocessing stage to avoid data leakage.
