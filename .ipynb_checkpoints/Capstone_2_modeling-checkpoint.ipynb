```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import joblib

from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint, uniform


from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, root_mean_squared_error
```

```python

X_train = pd.read_parquet("X_train_trimmed_encoded.parquet")
X_test = pd.read_parquet("X_test_encoded.parquet")
y_train = pd.read_parquet("y_train_trimmed.parquet").squeeze()
y_test = pd.read_parquet("y_test.parquet").squeeze()
```

```python
#Even though we did drop the 'rounded_averages' in order to avoid leaks, it still came up.
X_train = X_train.drop('rounded_averages', axis=1, errors='ignore')
X_test = X_test.drop('rounded_averages', axis=1, errors='ignore')
```

```python
#Although we did not see a correlation between the voting average and other variables,
#we start with linear regression then move onto decision tree.
LiReg = LinearRegression()
LiReg.fit(X_train,y_train)
```

```python
y_pred = LiReg.predict(X_test)

print("R-squared Score:", r2_score(y_test,y_pred))
print("MAE:", mean_absolute_error(y_test,y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
#Importing decision tree, randomforest regressor, and gradientboosting regressor.
dt_model = DecisionTreeRegressor(random_state=42)
```

```python
dt_model.fit(X_train,y_train)
```

```python
y_pred_dt = dt_model.predict(X_test)
```

```python
print('Decision Tree Performance:')
print('R-squared :', r2_score(y_test, y_pred_dt))
print('MAE :', mean_absolute_error(y_test, y_pred_dt))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_dt)))
```

```python
#Creating the random forest model, since decision tree is doing worse than the linear regression model. We move onto other tree models.
rf_model = RandomForestRegressor(random_state=42)
```

```python
rf_model.fit(X_train, y_train)
```

```python
y_pred_rf = rf_model.predict(X_test)
```

```python
print('RandomForest Regressor Performance:')
print('R-squared :', r2_score(y_test, y_pred_rf))
print('MSE :', mean_absolute_error(y_test, y_pred_rf))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_rf)))
```

```python
#Testing with Gradient Boosting Model before fine tuning it.
gb_model = GradientBoostingRegressor(random_state=42)
```

```python
gb_model.fit(X_train,y_train)
```

```python
y_pred_gb = gb_model.predict(X_test)
```

```python
print('GradientBoosting Regressor Performance:')
print('R-squared :', r2_score(y_test, y_pred_gb))
print('MSE :', mean_absolute_error(y_test, y_pred_gb))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_gb)))
```

```python
#RandomForest was the best performing model so far, now we will fine tune it
#First we plot feature importances

importances = rf_model.feature_importances_
features = X_train.columns
feat_importance = pd.Series(importances, index = features).sort_values(ascending = False)
plt.figure(figsize = (10,8))
sns.barplot(x=feat_importance[:10], y=feat_importance.index[:10])
plt.title('Top 10 Feature importances(Random Forest)')
plt.xlabel('Importance')
plt.ylabel('Features')
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
#Defining Param-grid for GridSearchCV that will be used to tune random forest model.
param_grid = {
                'n_estimators': [100,200],
                'max_depth' : [10,20,40],
                'min_samples_split': [2,5],
                'min_samples_leaf': [1,2],
                'max_features' : ['log2','sqrt']}
            
```

```python
#Setting up GridSearchCV for fine tuning
grid_search = GridSearchCV(estimator= RandomForestRegressor(random_state=42),param_grid=param_grid, scoring='r2', cv=3, n_jobs=1, verbose =2)
```

```python
grid_search.fit(X_train, y_train)
```

```python
y_pred_gscv = grid_search.predict(X_test)
```

```python
print("R-squared:", r2_score(y_test, y_pred_gscv))
print("MAE:", mean_absolute_error(y_test, y_pred_gscv))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_gscv)))
```

```python
#Importing RandomizedSearchCV and randint
#Testing RandomizedGSearchCV

params = {
     'n_estimators': randint(100, 300),
    'max_depth': randint(10, 40),
    'max_features': ['sqrt', 'log2', None],
    'min_samples_split': randint(2, 10),
    'min_samples_leaf': randint(1, 5)
}

search = RandomizedSearchCV(
    estimator=RandomForestRegressor(random_state=42),
    param_distributions=params,
    n_iter=5,
    cv=3,
    scoring='r2',
    verbose=1,
    n_jobs=-1
)
random_search = RandomizedSearchCV(estimator= RandomForestRegressor(random_state=42), param_distributions=params, scoring='r2', cv=3,n_jobs=1,verbose=2)
random_search.fit(X_train,y_train)
```

```python
y_pred_rdcv = random_search.predict(X_test)
#Print the results
print("R-squared:", r2_score(y_test, y_pred_rdcv))
print("MAE:", mean_absolute_error(y_test, y_pred_rdcv))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_rdcv)))
```

```python
# Clean column names to remove brackets which are not allowed in XGBoost
X_train.columns = [col.replace('[', '_').replace(']', '') for col in X_train.columns]
X_test.columns = [col.replace('[', '_').replace(']', '') for col in X_test.columns]
```

```python
##Installed Xgboost using pip install.
xgboost = XGBRegressor(random_state = 42)
xgboost.fit(X_train, y_train)
```

```python
y_pred_xgb = xgboost.predict(X_test) 
```

```python
print("R-squared:", r2_score(y_test, y_pred_xgb))
print("MAE:", mean_absolute_error(y_test, y_pred_xgb))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_xgb)))
```

```python
#XGBoost is the best model till now, we will fine tune it now.
#We also imported 'uniform' from scipy, so that we dont have fixed values in learning rate.
params = {
         'n_estimators':randint(100,500),
         'max_depth':randint(3,10),
         'learning_rate':uniform(0.01,0.3),
         'subsample':[0.6,0.8,1.0],
         'colsample_bytree':[0.6,0.8,1.0]
}
#Forcing CPU usage due to CUDA errors

xgb = XGBRegressor(random_state=42,tree_method ='hist')
rand_search = RandomizedSearchCV(
              estimator=xgb,
              param_distributions=params,
              n_iter=20,
              scoring= 'r2',
              cv= 3,
              n_jobs=1,
              verbose=2
)

rand_search.fit(X_train,y_train)

#Best Model

best_model = rand_search.best_estimator_
y_pred_tune = best_model.predict(X_test)

#Print the results

print('R2:', r2_score(y_test, y_pred_tune))
print('MAE:', mean_absolute_error(y_test, y_pred_tune))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_tune)))
```

```python
#Creating a dataframe to save all metrics for the models.

model_metrics=pd.DataFrame([
    {
        'Model':'Linear Regression',
        'R2': 0.19,
        'MAE':0.54,
        'RMSE':0.78
    },
    {
        'Model':'Decision Tree',
        'R2': 0.08,
        'MAE':0.63,
        'RMSE':'0.79'
    },
    {
        'Model':'Random Forest',
        'R2': 0.42,
        'MAE':0.48,
        'RMSE':0.67
    },
    {
        'Model':'Gradient Boosting',
        'R2': 0.40,
        'MAE':0.50,
        'RMSE':0.68
    },
    {
        'Model':'XGBoost Randomized Tuned',
        'R2': 0.46,
        'MAE':0.47,
        'RMSE':0.64
    }
])

#Displaying the results
model_metrics.sort_values(by='R2', ascending=False,inplace =True)
display(model_metrics)
```

```python
#Using the best model to make final predictions on X_test.

final_pred = best_model.predict(X_test)

#Creating DataFrame to compare predictions vs actuals.
pred_check= pd.DataFrame({'Actual Voting Average':y_test,
                          'Predicted Voting Average':final_pred
                         })
#Printing first 10 predictions using head() method.
print(pred_check.head())
```

```python
model_metrics.to_csv('model_metrics.csv', index = False)
pred_check.to_csv('pred_check.csv', index = False)
```
