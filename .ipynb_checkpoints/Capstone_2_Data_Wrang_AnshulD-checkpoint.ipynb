```python
#Import Libraries that we require

import pandas as pd
import numpy as np
import ast 
from library.sb_utils import save_file 

tmdb_reviews = pd.read_csv('tmdb10kmovies.csv')
tmdb_reviews.head()
```

```python
#Dropping Unnamed, backdrop_path, and poster_path Columns because they will be of no use in modeling.
tmdb_clean= tmdb_reviews.drop(columns=['Unnamed: 0', 'poster_path', 'backdrop_path'])
tmdb_clean.head()
```

```python
tmdb_clean.dtypes
```

```python
#Converting release_date dtype from object to datetime

tmdb_clean['release_date'] = pd.to_datetime(tmdb_clean['release_date'], errors= 'coerce')
tmdb_clean.dtypes
```

```python
#To Check whether there are any missing values in the dataset

missing_val = tmdb_clean.isnull().sum().sort_values(ascending = False)
print(missing_val)
```

```python
#Keeping 'title' as identifier for interpretation not as a model feature.
#Dropping 'tagline', 'imdb_id', and 'overview' columns because we will not be using it for modeling.

tmdb_clean = tmdb_clean.drop(columns = ['imdb_id', 'tagline','overview'])
tmdb_clean.head()

```

```python
tmdb_clean.dtypes
```

```python
# Setting id column as index
tmdb_clean = tmdb_clean.set_index('id')
```

```python
#Setting title column as first column for better readibility
cols = ['title'] + [col for col in tmdb_clean.columns if col != 'title']
tmdb_clean = tmdb_clean[cols]
tmdb_clean.head()
```

```python
#Quick check to see if we missed something missing
missing_val = tmdb_clean.isnull().sum().sort_values(ascending = False)
print(missing_val)
```

```python
tmdb_clean = tmdb_clean.dropna(subset = ['release_date'])
```

```python
tmdb_clean.head(20)
```

```python
#Checking if the list of actors and genres are strigified lists
type(tmdb_clean['genres'].iloc[0])
```

```python
type(tmdb_clean['actors'].iloc[0])
```

```python
#Parsing stringified lists column to list, by importing ast
for col in ['genres', 'actors', 'keywords', 'director']:
    tmdb_clean[col] = tmdb_clean[col].apply(lambda x: ast.literal_eval(x) if isinstance (x,str) else x)

type(tmdb_clean['genres'].iloc[0])
type(tmdb_clean['actors'].iloc[0])
```

```python
#The Stringified lists have already been parsed above.
#Using the explode() function to turn the list in the cell into multiple rows.
#We will group them later in pre processing, to make more features like avg_popularity.

actors_exploded = tmdb_clean.explode('actors')
#Cleaning actor names to get rid of unnecessary space 
actors_exploded['actors'] = actors_exploded['actors'].str.strip()
```

```python
#Final check to see if we have any missing values left
tmdb_clean.isna().sum()
```

```python
#Final check on dtypes to make sure everything is in order
tmdb_clean.dtypes
```

```python
#saving files
tmdb_clean.to_csv('tmdb_clean.csv')
```

```python
#Summary of Notebook
In this project, the TMDb dataset was cleaned and structured to prepare it for further analysis and modeling.
The workflow began with the removal of unnecessary columns and the conversion of release_date from string to datetime format to ensure consistency. 
Missing values in critical columns were identified and handled appropriately.
Several columns such as genres, actors, keywords, and director contained stringified lists,
which were parsed into native Python objects using ast.literal_eval() to make them usable for downstream processing.
The dataset was reorganized for better readability by setting id as the index and placing title as the first column. 
These steps completed the data wrangling process, resulting in a clean and structured dataset, 
ready for feature engineering and exploratory data analysis in the subsequent phases.

```
